```python
def impute_day(ranks,week,day=1, missing_fraction=0.2):
    assert week in [1,2,3]
    if week==1:
        W = csv2tensor.WEEK1
    elif week==2:
        W = csv2tensor.WEEK2
    elif week==3:
        W = csv2tensor.WEEK3
        
    wdtensor = W[day][0]
    #wdtensor = tl.tensor(wdtensor)
    random_mask = imputer.random_mask(wdtensor, missing_fraction)
    
    recErr, impErr = [],[]
    
    for R in ranks:
        a,b,c,re, ie = imputer.impute(wdtensor, rank=R, tol=1e-10,  n_iters=10000, mask=random_mask)
        
        recErr.append(re[-1])
        impErr.append(ie[-1])
        
    return recErr,impErr
```

```python
def impute_all_days(ranks, week, missing_fraction=0.1):
    daysRE,daysIE=[],[]
    for day in range(7):
        re,ie = impute_day(week=week,day=day, ranks=ranks, missing_fraction=missing_fraction)
        daysRE.append(re)
        daysIE.append(ie)
        
    return daysRE, daysIE
```

```python
def tl_all_days_mean(ranks, week):
    assert week in [1,2,3]
    if week==1:
        W = csv2tensor.WEEK1
    elif week==2:
        W = csv2tensor.WEEK2
    elif week==3:
        W = csv2tensor.WEEK3
        
    
    
    TLdaysRE=[]
    for R in ranks:
        print(R,end=' ')
        total_re = []
        for d in range(7):
            (w,f),e = td.parafac(W[d][0], rank=R, return_errors=True)
            total_re.append(e[-1])
            
        TLdaysRE.append(np.array(total_re).mean()) #mean
        
    return TLdaysRE
```

```python
ranks = list(range(1,100,2))
missing_fractions = np.arange(0.1,1, 0.1)

results = []

for mf in missing_fractions:
    daysRE, daysIE = impute_all_days(ranks, week=1,missing_fraction=mf)
    results.append((daysRE,daysIE)) 
```

```python
TLdaysRE = tl_all_days_mean(ranks=ranks,week=1)
```

```python
sns.set_theme()
fig, axes = plt.subplots(4, 2, sharex=True,figsize=(16,20))
for mf,result,idx in zip(missing_fractions,results, range(len(results))):
    
    if mf < 0.9:
        re,ie = np.array(result[0]),np.array(result[1])
        re,ie = re.mean(axis=0), ie.mean(axis=0)
        
        thisax = axes[idx//2][idx%2]
        
        sns.lineplot(ax = thisax , x=ranks, y = re, color='orange', label=f'rec')
        sns.lineplot(ax = thisax , x=ranks, y = ie, color='green', label=f'imp')
        sns.lineplot(ax=thisax, x=ranks, y=TLdaysRE, color='black', label='tensorly cp [no missing]')
        
        thisax.set_title(f'missing = {mf*100:.0f}%')
        
fig.suptitle("Rank vs normalized MSE Rec/Imp error [avg. over 7 days of week2]")
```
